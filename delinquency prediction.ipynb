```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix, roc_curve, auc
from sklearn.preprocessing import StandardScaler
```

```python
# Monthly payment columns
month_cols = ['Month_1', 'Month_2', 'Month_3', 'Month_4', 'Month_5', 'Month_6']

# Create behavior summaries
df['Late_Count'] = df[month_cols].apply(lambda row: sum(row == 'Late'), axis=1)
df['Missed_Count'] = df[month_cols].apply(lambda row: sum(row == 'Missed'), axis=1)
df['OnTime_Count'] = df[month_cols].apply(lambda row: sum(row == 'On-time'), axis=1)

# Drop month columns and ID
df.drop(columns=month_cols + ['Customer_ID'], inplace=True)
```

```python
for col in ['Income', 'Credit_Score', 'Loan_Balance']:
    df[col].fillna(df[col].median(), inplace=True)

#Median imputation is robust and avoids dropping data unnecessarily.
```

```python
# Encode Categorical Variables
df = pd.get_dummies(df, columns=['Employment_Status', 'Credit_Card_Type', 'Location'], drop_first=True)

## Logistic regression needs numeric inputs; one-hot encoding allows model to interpret categories.
```

```python
# Define Features and Target
X = df.drop('Delinquent_Account', axis=1)
y = df['Delinquent_Account']

## Separating X and y is needed for model training.

```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)

##  Stratified split keeps the same proportion of 0/1 in both sets to avoid bias.

```

```python
# Scale Features - Logistic regression assumes normally scaled inputs for optimal performance.
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# Train Logistic Regression
model = LogisticRegression(class_weight='balanced', max_iter=1000, random_state=42)
model.fit(X_train_scaled, y_train)
```

```python
# Evaluate Performance
y_pred = model.predict(X_test_scaled)
print("Classification Report:\n", classification_report(y_test, y_pred))
print("Confusion Matrix:\n", confusion_matrix(y_test, y_pred))
## These metrics help assess accuracy, recall, and fairness across classes.
```
